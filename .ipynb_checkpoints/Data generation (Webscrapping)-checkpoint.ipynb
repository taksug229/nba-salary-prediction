```python
# use findALL() to get the column headers
soup.findAll('tr', limit=2)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers = headers[1:]
headers
```

```python
soup.findAll('tr')[1:3]
```

```python
# avoid the first header row
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
type(player_stats)
```

```python
df_stats = pd.DataFrame(player_stats, columns=headers)
df_stats.dropna(inplace=True)
df_stats.drop_duplicates('Player', keep='first', inplace=True)
df_stats = df_stats.reset_index(drop=True)
```

```python
# Converted Special Characters to unicode and removed bad words and characters.

s = df_stats['Player']
coded = []
for i in s:
    coded.append(unidecode(u'{}'.format(i)))

bad_chars = [".", "'"]
string_list = []
for word in coded:
    string = word.replace(' Jr.', '')
    string = string.replace(' III', '')
    string = ''.join(i for i in string if not i in bad_chars) 
    string_list.append(string)
df_stats['Player'] = pd.Series(string_list)
```

```python
# Scrape Salary information

# NBA season we will be analyzing
year = '2018-2019'
# URL page we will scraping (see image above)
url = "https://hoopshype.com/salaries/players/{}/".format(year)
# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
# use findALL() to get the column headers
soup.findAll('tr', limit=1)
# use getText()to extract the text we need into a list
headers = [(th.getText().strip()) for th in soup.findAll('tr', limit=1)[0].findAll('td')]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers[0] = 'Rank'
```

```python
soup.findAll('tr', limit=2)
```

```python
soup.findAll('tr')[1:]
```

```python
# avoid the first header row
rows = soup.findAll('tr')[1:]
player_salary = [[td.getText().strip() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
df_salary = pd.DataFrame(player_salary, columns = headers)
```

```python
df_salary['Player']
```

```python
# Cleaning Salary Name

names = df_salary['Player']

coded_names = []
for i in names:
    coded_names.append(unidecode(u'{}'.format(i)))

bad_chars = [".", "'"]
salary_list = []
for word in coded_names:
    string = word.replace(' Jr', '')
    string = string.replace(' III', '')
    string = ''.join(i for i in string if not i in bad_chars) 
    salary_list.append(string)
```

```python
df_salary['Player'] = pd.Series(salary_list)
```
